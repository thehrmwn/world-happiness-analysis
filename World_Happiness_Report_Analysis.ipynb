## üëã **Introduction**

üîé **Context**

The World Happiness Report is a landmark survey of the state of global happiness. The first report was published in 2012, the second in 2013, the third in 2015, and the fourth in the 2016 Update. The World Happiness 2017, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th. The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields ‚Äì economics, psychology, survey analysis, national statistics, health, public policy and more ‚Äì describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

üìë **Content**

The happiness scores and rankings use data from the Gallup World Poll. The scores are based on answers to the main life evaluation address inquired within the survey. This address, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. The scores are from nationally representative samples for the years 2013-2016 and use the Gallup weights to make the estimates representative. The columns following the happiness score estimate the extent to which each of six factors ‚Äì economic production, social support, life expectancy, freedom, absence of corruption, and generosity ‚Äì contribute to making life evaluations higher in each country than they are in Dystopia, a hypothetical country that has values equal to the world‚Äôs lowest national averages for each of the six factors. They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.

The Happiness Score is a national average of the responses to the main life evaluation question asked in the Gallup World Poll (GWP), which uses the Cantril Ladder.

üí® **About the Dataset** (*that will i use*)

- **GDP per capita** ‚û° measure of a country's economic output that accounts for its number of people
- **Healthy Life Expectancy** ‚û° the average number of years that a newborn can expect to live in "full health"‚Äîin other words, not hampered by disabling illnesses or injuries.
- **Social support** ‚û° having friends and other people, including family, to turn to in times of need or crisis to give you a broader focus and positive self-image. Social support enhances quality of life and provides a buffer against adverse life events.
- **Freedom to make life choices** ‚û° an individual's opportunity and autonomy to perform an action selected from at least two available options, unconstrained by external parties
- **Generosity** ‚û° the quality of being kind and generous.
- **Corruption Perception** ‚û° The Corruption Perceptions Index (CPI) is an index published annually by Transparency International since 1995 which ranks countries "by their perceived levels of public sector corruption, as determined by expert assessments and opinion surveys. The scale from 100 (very clean) to 0 (highly corrupt) [CPI source](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)

#### **Happiness Score**

Correlation between Happiness Score with other variables.

```python
y_column = 'Happiness score'
x_columns = ['GDP', 'Social Support', 'Healthy Life', 'Freedom', 'Generosity', 'Perceptions of Corruption']
fig, axes = plt.subplots(2, 3, figsize=(12, 10))
axes = axes.flatten()

for i, x_col in enumerate(x_columns):
    ax = axes[i]
    ax.scatter(y = data['Happiness score'], x = data[x_col], alpha=0.5)
    ax.set_title(f'Happiness vs {x_col}', fontsize=14)
    ax.set_xlabel(x_col, fontsize=12)
    ax.set_ylabel(y_column, fontsize=12)

plt.tight_layout()
plt.show()
```

It shows that, every variable has correlation with Happiness Score:
1. Variable GDP, Social Support, Healthy Life, and Freedom. **the higher the variable is in a country, the higher the country's happiness level.**
2. Generosity is not to much affect of Happiness Level, because sometimes **a country that has average happiness level more Generous than a country that have higher levels of happines**s.
3. Perceptions of Corruption. It will affect the Happiness Level of a country. **The higher the PCI value of a country, the more people will trust the government**. And vice-versa.
